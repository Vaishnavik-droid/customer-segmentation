## data cleaning and preprocessing

```python
import pandas as pd
import numpy as np
```

```python
df=pd.read_csv("Airbnb_Open_Data.csv",low_memory=False) #license column contains different types of values,so we have to use low_mwemory=false.
```

```python
#view the dataset
df.head()
```

```python
#check dataset information
df.info()
```

```python
df.shape # column and rows
```

```python
#remove extra spaces column name
df.columns=df.columns.str.strip()
```

```python
#check missing values
df.isnull().sum()
```

```python
df.describe() #showing statistical summary 
```

```python
#remove duplicates
df.duplicated().sum()
```

```python
#remove duplicate records
df=df.drop_duplicates()
```

### handle missing values

```python
# most missing values involves license column so drop it.
df.drop(columns=["license"],inplace=True)
```

```python
df.drop(columns=["NAME"],inplace=True) #drop the name column
```

```python
#check missing values
df['host_identity_verified'].isnull().sum()
```

```python
df['host_identity_verified'].unique()
```

```python
df['host_identity_verified']=df['host_identity_verified'].fillna(df['host_identity_verified'].mode()[0])
```

```python
#verify 
df['host_identity_verified'].isnull().sum()
```

```python
#check the missing value remains
df['host name'].isnull().sum()
```

```python
df['host name']=df['host name'].fillna('Unknown')
```

```python
#verify
df['host name'].isnull().sum()
```

```python
#.filna= replace all NaN value with the value you provide 
#.mode()return the most frequent value in that columns,it returns the a series ,so [0] selects the the first mode 
df['neighbourhood group']=df['neighbourhood group'].fillna(df['neighbourhood group'].mode()[0])
```

```python
df.drop(columns=["neighbourhood"],inplace=True)
```

```python
#fill missing with mode
df['country']=df['country'].fillna('unknown')
```

```python
df['country code']=df['country code'].fillna(df['country code'].mode()[0])
```

```python
df['instant_bookable']=df['instant_bookable'].fillna(df['instant_bookable'].mode()[0])
```

```python
df['room type']=df['room type'].fillna(df['room type'].mode()[0])
```

```python
#fill missing with median
df['Construction year']=df['Construction year'].fillna(df['Construction year'].median())
```

```python
df['price'].dtype # this columns contain string
```

```python
df['price'].head(10)
```

```python
#convert into numeric
df['price']=pd.to_numeric(df['price'].replace(r'[\$]','',regex=True),errors='coerce')
```

```python
df['price'].dtype
```

```python
#fill the missing values
df['price']=df['price'].fillna(df['price'].median())
```

```python
df['price'].isnull().sum()
```

```python
#remove $ and convert the numeric using regex - r'[\$']->removes $ and ,
#errors='coerce' is the parameter of pd.to_numeric()
df['service fee'] = pd.to_numeric( df['service fee'].replace(r'[\$,]', '', regex=True),errors='coerce')

```

```python
#fill the missing values
df['service fee']=df['service fee'].fillna(df['service fee'].median())
```

```python
df['service fee'].isnull().sum()
```

```python
df['minimum nights']=df['minimum nights'].fillna(df['minimum nights'].median())
```

```python
#errors='coerce' is also parameter  of pd.to_datetime
df['last review']=pd.to_datetime(df['last review'],errors='coerce')
```

```python
#fill the missing values
df['last review']=df['last review'].fillna(df['last review'].mode()[0])
```

```python
df['last review'].isnull().sum()
```

```python
df['reviews per month']=df['reviews per month'].fillna(df['reviews per month'].median())
```

```python
df['review rate number']=df['review rate number'].fillna(df['review rate number'].median())
```

```python
df['calculated host listings count']=df['calculated host listings count'].fillna(df['calculated host listings count'].median())
```

```python
df['availability 365'].dtype # this columns contain object
```

```python
#check what values in the col
df['availability 365'].head(20)
```

```python
#if the values are just numbers stored as string,run:
df['availability 365']=pd.to_numeric(df['availability 365'],errors='coerce')
```

```python
df['availability 365']=df['availability 365'].fillna(df['availability 365'].median())
```

```python
df['availability 365'].head(10)
```

```python
df['house_rules']=df['house_rules'].fillna('Not provided')
```

```python
df.to_csv("Airbnb_Open_Data_cleaned.csv",index=False)
```
