### 11.Price vs Minimum Nights
This investigates the relationship between **listing price** and **minimum nights reqiured for booking**.
It helps determine whether the hosts requiring longer minimum stays tend to charge higher or lower rates and helps to identify any potential relationship between booking duration and cost.

```python
bivariate_df = df[(df["minimum nights"] >= 0) & (df["minimum nights"] <= 365)]
#Filtering the dataset to include only listings with minimum nights between 0 and 365 for better visualization.
#Using a scatter plot for visualization.
plt.figure(figsize=(10,6))
sns.scatterplot(
    data=bivariate_df,
    x="minimum nights",
    y="price",
    alpha=0.4
)
plt.title("Price vs. Minimum Nights")
plt.xlabel("Minimum Nights")
plt.ylabel("Listing Price")
plt.xlim(0,100)
plt.show()
```

### Observation:
The scatter plot indicates no clear relationship between listing price and minimum nights. Prices remain widely distributed across all minimum-night requirements, suggesting that listings requiring longer stays are not consistently more expensive than those permitting shorter bookings. Most observations are concentrated between 1 and 10 minimum nights, reflecting the dominance of short-term rental policies on the platform. Overall, the analysis suggests that minimum stay requirements have little observable influence on listing prices within this dataset.

### 13. Correlation Analysis of Numerical Features.
Correlation analysis measures the strength and direction of the linear relationship between the numeric variables, A correlation heatmap provides us with a visual summary of this relation and helps us identify highly related features and independant variables and potential redundancy before performing customer segmentation.

```python
#we select the valuable numeric columns for correlation analysis , we exclude identity column as it just identifier.
numeric_cols =["price", "service fee", "minimum nights","number of reviews","reviews per month","review rate number",
            "calculated host listings count",
            "availability 365"]
#the corr() method computes the correlation matrix.
corr = df[numeric_cols].corr()

#Heatmap of the correlation matrix
plt.figure(figsize=(10,8))
sns.heatmap(corr,
            annot=True,
            cmap="coolwarm",
            fmt=".2f",
            linewidths=0.5,
            square=True,)
plt.title("Correlation Heatmap of Numerical Features")

plt.show()
```

### Observation:
The correlation analysis indicates that price and service fee exhibit the strongest positive relationship (r = 0.75), suggesting that higher-priced listings generally incur higher service fees. A moderate positive correlation (r = 0.55) is also observed between number of reviews and reviews per month, indicating that frequently reviewed listings tend to accumulate a larger number of reviews over time. Apart from these relationships, most numerical features display very weak or negligible correlations, implying that they contribute largely independent information. No significant negative correlations are present in the dataset.
