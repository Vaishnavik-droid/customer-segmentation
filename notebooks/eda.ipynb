## Univariate Analysis
### 1.Price
Price is an important numeric feature that will help us identify pricing trends, detect outliers and understand how listings are priced.

```python
print(df["price"].describe())
print(f"Skewness : {df['price'].skew():.2f}")
```

```python
plt.figure(figsize=(10,5))

sns.histplot(df["price"],bins=30,kde=True)

plt.title("Distribution of Listing Prices")
plt.xlabel("Price")
plt.ylabel("Frequency")

plt.show()
```

```python
plt.figure(figsize=(10,2))
sns.boxplot(x=df["price"])
plt.title("Boxplot of Listing Prices")
plt.show()
```

The price is distributed around 500-550,i.e large frequency of listings are priced in this range. The histogram exhibits a pronounced peak around median price, while overall dist. is symmetric as confirmed by skewness close to zero.The boxplot did not indicate presence of outliers, suggesting extreme prices have been largely dropped in preprocessing,the interquartile range is relatively wide, indicationg noticible variation in listing prices across the dataset.

```python
df["price"].value_counts().head(10)
```

## Important!
A significant spike was observed at price value of 524, appearing approx 17.7% of the dataset. Such concentration of a single price variable is highly unlikely in a real-world airbnb dataset and suggest that the dataset may have imputed or synthetic values/records.Consider this while interpreting the results of subsequent analyses and the customer segmentation.

```python
#Checking whether service fee is derived from price or not.
df[["price","service fee"]].corr()
```

Since the correlation isnt extremely high (0.90+) we shall analyse service fee as it has enough independent variation and is financially meaningful feature.

### 3.Minimum Nights.
The **minimum nights** feature represents the minimum number of nights a guest must have to book in order to reserve a listing.This feature will help us understand the booking policies set by host and provide insights into types of stay offered.Analysing its distribution may also reveal unsual book requirements.

```python
print(df["minimum nights"].describe())
print(f"Skewness : {df['minimum nights'].skew():.2f}")
```

```python
plt.figure(figsize=(10,5))

sns.histplot(df["minimum nights"],bins=30,kde=True)

plt.title("Distribution of Minimum Nights")
plt.xlabel("Minimum Nights")
plt.ylabel("Frequency")

plt.show()
```

```python
print(df["minimum nights"].min())
print(df["minimum nights"].max())
```

To improve the readability of the histogram, extreme outliers (-1223 and 5645) are excluded **only for visualization purposes**. These values are retained in the original dataset and are discussed separately as data quality anomalies.

```python
minimum_nights_fil =df[(df["minimum nights"]>=0)& (df["minimum nights"]<=365)]["minimum nights"]
```

```python
plt.figure(figsize=(10,5))
sns.histplot(minimum_nights_fil,discrete=True)
plt.title("Distribution of Minimum Nights (0-365)")
plt.xlabel("Minimum Nights")
plt.ylabel("Frequency")
plt.show()
```

```python
df["minimum nights"].value_counts().head(10)
```

```python
plt.figure(figsize=(10,2))

sns.boxplot(
    x=minimum_nights_fil
)

plt.title("Boxplot of Minimum Nights (0–365 Days)")
plt.xlabel("Minimum Nights")

plt.show()
```

```python
df[(df["minimum nights"] < 0) | (df["minimum nights"] > 365)][["minimum nights"]]
```

#### Observations:
1.The minimum night feature is heavily concentrated between 1 and 5 nights with a median of 3 nights.Indicating that most of the listings are meant for short stays.
2.A secondary noticeable peak at 30 nights suggests presence of long-term rental listings.
3.The distribution is extremely right skewed (skewness=89.42)due to a small number of extreme outliers.
4.The boxplot reveals numerous high value outliers, while two invalid observations(-1223,5645) were identified as data qaulity anomalies and excluded via filtering from the visualization to improve readability.
